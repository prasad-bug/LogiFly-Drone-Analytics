## 💰 SECTION A — Business & Financial Analysis
### Question 1: Calculate Total Annual Savings (35% cost reduction)
### Question 2: Calculate Net Benefit = Total Savings – Investment

```python
# ── Given Data ──────────────────────────────────────────────────
ANNUAL_COST   = 1_50_00_000   # ₹1.5 Crore
INVESTMENT    = 85_00_000     # ₹85 Lakhs
REDUCTION_PCT = 0.35          # 35%

# ── Q1: Annual Savings ──────────────────────────────────────────
annual_savings = ANNUAL_COST * REDUCTION_PCT
print('=' * 55)
print('  QUESTION 1 — Total Annual Savings')
print('=' * 55)
print(f'  Annual Operational Cost  : ₹{ANNUAL_COST:>14,.0f}')
print(f'  Cost Reduction           :          35.00%')
print(f'  Annual Savings (35%)     : ₹{annual_savings:>14,.0f}')
print(f'  = ₹{annual_savings/100000:.2f} Lakhs per year')

# ── Q2: Net Benefit ─────────────────────────────────────────────
net_benefit = annual_savings - INVESTMENT
payback_months = INVESTMENT / (annual_savings / 12)
roi_pct = (net_benefit / INVESTMENT) * 100

print()
print('=' * 55)
print('  QUESTION 2 — Net Benefit')
print('=' * 55)
print(f'  Formula: Net Benefit = Total Savings – Investment')
print(f'  Net Benefit (Year 1) : ₹{annual_savings:,.0f} – ₹{INVESTMENT:,.0f}')
print(f'                       = ₹{net_benefit:>+,.0f}')
print(f'  = ₹{net_benefit/100000:.2f} Lakhs (negative in Year 1 = recovery phase)')
print()
print(f'  Payback Period       : {payback_months:.1f} months (~{payback_months/12:.1f} years)')
print(f'  ROI (Year 1)         : {roi_pct:.1f}%')
print()
```

```python
# 5-year projection
years, savings_list, investments, net_benefits, cumulative = [], [], [], [], []
cum = 0
for y in range(1, 6):
    s = ANNUAL_COST * REDUCTION_PCT * (1 + 0.03 * (y - 1))
    i = INVESTMENT if y == 1 else INVESTMENT * 0.05
    nb = s - i
    cum += nb
    years.append(f'Year {y}')
    savings_list.append(round(s))
    investments.append(round(i))
    net_benefits.append(round(nb))
    cumulative.append(round(cum))

df_roi = pd.DataFrame({
    'Year': years,
    'Annual Savings (₹)': savings_list,
    'Investment (₹)': investments,
    'Net Benefit (₹)': net_benefits,
    'Cumulative Net (₹)': cumulative
})
print('5-Year Financial Projection:')
print(df_roi.to_string(index=False))
print(f'\nTotal 5-Year Net Benefit: ₹{sum(net_benefits):,.0f} = ₹{sum(net_benefits)/100000:.1f} Lakhs')
```

```python
# Plot ROI
fig, axes = plt.subplots(1, 2, figsize=(16, 6), facecolor=BG)

# Left: Annual cash flow
ax1 = axes[0]
ax1.set_facecolor(PANEL)
x = np.arange(5)
w = 0.3
bars_s = ax1.bar(x - w/2, [s/100000 for s in savings_list], w, label='Annual Savings', color=BLUE, alpha=0.85)
bars_i = ax1.bar(x + w/2, [i/100000 for i in investments],  w, label='Investment/Maint.', color=RED, alpha=0.8)
ax1.plot(x, [nb/100000 for nb in net_benefits], 'o-', color=YELLOW, lw=2.5, ms=8, label='Net Benefit', zorder=5)
for xi, nb in zip(x, net_benefits):
    ax1.annotate(f'₹{nb/100000:.1f}L', (xi, nb/100000), textcoords='offset points',
                 xytext=(0, 10), ha='center', fontsize=9, color=YELLOW, fontweight='bold')
ax1.axhline(0, color='white', lw=0.8, ls='--', alpha=0.4)
ax1.set_xticks(x); ax1.set_xticklabels(years, color='lightgray')
ax1.set_ylabel('₹ Lakhs', color='lightgray')
ax1.set_title('Annual Cash Flow', color='white', fontsize=13, fontweight='bold')
ax1.legend(facecolor='#0d1b2a', edgecolor='#1e3a5f', labelcolor='white')
ax1.tick_params(colors='lightgray'); ax1.spines[:].set_color('#1e3a5f')
ax1.grid(axis='y', alpha=0.2)

# Right: Cumulative net benefit
ax2 = axes[1]
ax2.set_facecolor(PANEL)
cumulative_plot = [c/100000 for c in cumulative]
ax2.fill_between(x, cumulative_plot, alpha=0.15, color=GREEN)
ax2.plot(x, cumulative_plot, 'o-', color=GREEN, lw=3, ms=10, zorder=5)
for xi, c in zip(x, cumulative):
    ax2.annotate(f'₹{c/100000:.1f}L', (xi, c/100000), textcoords='offset points',
                 xytext=(0, 12), ha='center', fontsize=9, color='lightgreen', fontweight='bold')
ax2.axhline(0, color=RED, lw=1.5, ls='--', alpha=0.7, label='Break-even line')
ax2.set_xticks(x); ax2.set_xticklabels(years, color='lightgray')
ax2.set_ylabel('₹ Lakhs', color='lightgray')
ax2.set_title('Cumulative Net Benefit', color='white', fontsize=13, fontweight='bold')
ax2.legend(facecolor='#0d1b2a', edgecolor='#1e3a5f', labelcolor='white')
ax2.tick_params(colors='lightgray'); ax2.spines[:].set_color('#1e3a5f')
ax2.grid(axis='y', alpha=0.2)

plt.suptitle('LogiFly Smart Drone — 5-Year Financial Projection', color='white', fontsize=15, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('/mnt/user-data/outputs/financial_projection.png', dpi=150, bbox_inches='tight', facecolor=BG)
plt.show()
print('Chart saved.')
```

### Question 4: Generate Synthetic Warehouse Data Using Python

```python
# ── Q4: Synthetic Data Generation ──────────────────────────────
n_days = 365
dates  = pd.date_range('2024-01-01', periods=n_days, freq='D')
improvement = np.linspace(0, 1, n_days)  # drone AI learns over time

# Manual system — noisy, inconsistent
manual_items    = np.random.normal(1800, 200, n_days).astype(int).clip(800, 2500)
manual_accuracy = np.random.normal(78, 5, n_days).clip(60, 90)
manual_errors   = np.random.normal(12, 3, n_days).clip(3, 25)
manual_time     = np.random.normal(6.5, 0.8, n_days).clip(4, 10)
manual_sat      = np.random.normal(62, 7, n_days).clip(40, 80)

# Drone system — high performance, improving over time
drone_items    = np.random.normal(4500, 100, n_days).astype(int).clip(3800, 5500)
drone_accuracy = np.clip(np.random.normal(96, 1.5, n_days) + improvement * 1.5, 90, 99.9)
drone_errors   = np.clip(np.random.normal(3, 0.8, n_days) - improvement * 0.5, 0.5, 6)
drone_time     = np.clip(np.random.normal(1.2, 0.15, n_days) - improvement * 0.2, 0.5, 2)
drone_sat      = np.clip(np.random.normal(88, 3, n_days) + improvement * 4, 80, 99)

df_wh = pd.DataFrame({
    'Date':                  dates,
    'Manual_Items_Scanned':  manual_items,
    'Drone_Items_Scanned':   drone_items,
    'Manual_Accuracy_Pct':   manual_accuracy.round(2),
    'Drone_Accuracy_Pct':    drone_accuracy.round(2),
    'Manual_Error_Rate_Pct': manual_errors.round(2),
    'Drone_Error_Rate_Pct':  drone_errors.round(2),
    'Manual_Scan_Time_Hr':   manual_time.round(2),
    'Drone_Scan_Time_Hr':    drone_time.round(2),
    'Manual_Satisfaction':   manual_sat.round(1),
    'Drone_Satisfaction':    drone_sat.round(1),
})
df_wh['Month'] = df_wh['Date'].dt.month
df_wh['Week']  = df_wh['Date'].dt.isocalendar().week.astype(int)

print('Synthetic Warehouse Dataset — Shape:', df_wh.shape)
print('\nFirst 5 rows:')
print(df_wh.head())
print('\nDescriptive Statistics:')
print(df_wh[['Manual_Items_Scanned','Drone_Items_Scanned',
             'Manual_Accuracy_Pct','Drone_Accuracy_Pct',
             'Manual_Error_Rate_Pct','Drone_Error_Rate_Pct']].describe().round(2))
```

## 🤝 SECTION C — CRM and Business Impact
### Question 5: How Inventory Errors Affect Client Satisfaction, Retention, and Revenue

```python
# Generate CRM synthetic data
np.random.seed(42)
n_clients = 120
sectors = ['E-Commerce','Retail','Manufacturing','Pharma','FMCG']

crm_rows = []
for i in range(n_clients):
    sector        = np.random.choice(sectors, p=[0.35, 0.25, 0.2, 0.1, 0.1])
    base_clv      = np.random.randint(5, 50) * 100000
    error_score   = np.random.uniform(0, 1)
    retention_pre = float(np.clip(np.random.normal(0.62, 0.12), 0.35, 0.82))
    retention_post= float(np.clip(retention_pre + np.random.uniform(0.18, 0.28), 0.65, 0.97))
    sat_pre       = float(np.clip(np.random.normal(58, 10), 35, 75))
    sat_post      = float(np.clip(sat_pre + np.random.normal(28, 6), 75, 99))
    complaints_pre = np.random.randint(5, 30)
    complaints_post= np.random.randint(1, 8)
    crm_rows.append({
        'Client_ID':         f'CLT{i+1:03d}',
        'Sector':            sector,
        'Base_CLV':          base_clv,
        'Error_Score':       round(error_score, 2),
        'Retention_Pre':     round(retention_pre, 3),
        'Retention_Post':    round(retention_post, 3),
        'Satisfaction_Pre':  round(sat_pre, 1),
        'Satisfaction_Post': round(sat_post, 1),
        'Complaints_Pre':    complaints_pre,
        'Complaints_Post':   complaints_post,
        'CLV_Pre':           int(base_clv * retention_pre),
        'CLV_Post':          int(base_clv * retention_post * 1.15),
    })

df_crm = pd.DataFrame(crm_rows)

print('CRM Synthetic Data — Shape:', df_crm.shape)
print(df_crm.head(8))
print('\nCRM Summary by Sector:')
print(df_crm.groupby('Sector')[['Satisfaction_Pre','Satisfaction_Post','Retention_Pre','Retention_Post']].mean().round(2))
```
